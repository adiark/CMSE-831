‚úçüèª  (10pts) From the standard penguins dataset, answer this question: what percentage of female penguins from Dream have a body mass greater than $4000$ grams? (Use only Pandas attributes and methods. Show your code as well as your answer.)


üìù **Answer**:

```python
# Loading libraries and dataset and performing analysis
from itertools import count

df_peng = sns.load_dataset("Penguins")
f_dream_data = df_peng.where(df_peng['sex'] == "FEMALE").where(df_peng['island'] == "Dream").dropna()
f_perc = len(f_dream_data.where(f_dream_data['body_mass_g'] > 4000).dropna())/len(f_dream_data)
print("Percentage of female penguins from dream with body mass greater than 4000 grams is",round(f_perc*100,2),"%")
```

## EDA: Seaborn and Correlations

‚úçüèª  (10pts) Read in the mpg dataset and perform EDA on it. Specifically:
* get the basic statistics from the df,
* make a `displot` of horsepower versus mpg, including a rug plot and with a kde,
* make a Seaborn `heatmap` of the correlations using Pandas' `.corr`, with annotations and your choice of colormap,
* make a `regplot` ([or, `lmplot`](https://seaborn.pydata.org/tutorial/regression.html)) of horsepower versus mpg using the lowess smoother; choose good colors,
* design a `FacetGrid` of your choice.

After each plot, open a markdown cell and give one sentence of what the plot tells you. That is, as a data scientist, what did this form of EDA reveal about the data? (If it didn't reveal anything to you, that is also an important possible answer.) 


üìù **Answer**:

```python
# Loading and describing mpg dataset
df_mpg = sns.load_dataset("mpg")
df_mpg.describe()
```

```python
# displot of horsepower vs mpg
sns.set_style("darkgrid")
sns.displot(data = df_mpg, x= "horsepower", y= "mpg", kind= "kde", rug= True).set(title = "Displot of horsepower vs mpg")
```

‚úèÔ∏è The chart tells us about the relation between horsepower and mpg, as visible less horsepower is proportionate to higher mpg 

```python
# heatmap of correlation values between attributes
sns.heatmap(df_mpg.corr(), annot= True,  cmap = "Blues").set(title = "Heatmap Correlation")
```

‚úèÔ∏è The heatmap tell us about how strongly attributes within dataset are correlated to eachother, help us filter dataset. 

```python
# regplot / lmplot of horsepower vs mpg using lowess smoother
sns.regplot(data= df_mpg, x= "horsepower", y= "mpg", lowess= True, color= "green").set(title= "Regression plot of horsepower vs mpg")
```

‚úèÔ∏è The line in above chart tells us an estimate relation between horsepower and mpg among the sample set, this type of regression is generally used in predicting continuous variables.

```python
# Facetgrid 
fig = sns.FacetGrid(data= df_mpg, col= "origin", hue= "cylinders")
fig.map(sns.scatterplot, "weight", "acceleration", alpha= 0.5)
fig.add_legend()
```

‚úèÔ∏è As clearly visible, we can see clusters of different colors revealing patterns, like four cylinders are optimum for a cars performance.

```python
# Facetgrid 
fig = sns.FacetGrid(data= df_mpg, col= "origin", hue= "cylinders")
fig.map(sns.scatterplot, "mpg", "weight", alpha= 0.5)
fig.add_legend()
```

‚úèÔ∏è As visible, more weight means less mpg.
