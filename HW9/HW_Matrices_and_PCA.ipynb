## Problem #3: Quadratic Form (20 pts)

In this problem you will explore a mathematical object that arises frequently in data science: the quadratic form. For now, let's just assume it is an interesting object and build our intuition; then, we'll put it use! 

The quadratic form is written as:
$${\bf x}^TA{\bf x}.$$
Here, ${\bf x}$ is a vector of arbitrary length and $A$ a matrix.


___
First question: if ${\bf x}$ has size $n\times 1$, what shape is $A$? What is the shape of ${\bf x}^T$? 

**Answer**: if ${\bf x}$ has size $n\times 1$, then shape of $A$ will be $n\times n$ and shape of ${\bf x}^T$ will ne $1\times n$
___

Ok, now let's work in 2D so that we can visualize what we are doing. Write ${\bf x}$ as
$${\bf x} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix},$$
where we think of the elements as axes in a 2D space. 

____
Second question: Let's assume that $A=I$. What is the quadratic form for this case - mathematical and meaning? 

**Answer**: If $A=I$ then the quadratic form becomes ${\bf x}^T{\bf x}$. This gives us the euclidean inner product or the sums of squares of the vectors $x_1$ and $x_2$ and is equal to $x_1^2 + x_2^2$. This means the resulting quadratic form is already aligned. Thus, when we plot the resulting curve we get an ellipse aligned parralel to the original coordinate system. 

Suppose you write this as ${\bf x}^T{\bf x} = c$ and plot the resulting curve in the $x_2-x_1$ space: what shape do you get? What role does $c$ play? How is the quadratic form itself connected to a [distance](https://en.wikipedia.org/wiki/Mahalanobis_distance)?

**Answer**: If we write ${\bf x}^T{\bf x} = c$, and plot the resulting curve we will get an ellipse parralel to coordinate system. The value of c represents the square distance of the quadratic form from the origin of the coordinate system. If we subtract mean from the vector and transform our coordinate system, then the distance c is called as squared Mahalanobis distance.
___
Great. Let's make $A$ more general now. 
___
Third question: For an arbitrary $A = \begin{bmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{bmatrix}$, write out the quadratic form explicitly. 

**Answer**: The quadratic form for an arbitrary A will be $a_{11}x_1^2 + (a_{12} + a_{21})x_1x_2 + a_{22}x_2^2$

Same as the previous question: if you plot the quadratic form in the $x_2-x_1$ space, what geometric shape do you get? What role do the diagonal elements play? What about the off-diagonal elements? 

**Answer**: If we plot the above quadratic form, we will get a shape resembling an ellipse with some rotation performed about the axes. The diagonal elements of the matrix $A$ will determine the length of the major and minor axis of the ellipse while the off diagonal elements controls the rotation and reflection of the ellipse about the $x_1$ and $x_2$ axes.   
_____

If you are not familiar with these geometric shapes, [read this](https://www.geometrictools.com/Documentation/InformationAboutEllipses.pdf). 

____
Fourth question: What if you [diagonalize](https://en.wikipedia.org/wiki/Eigendecomposition_of_a_matrix) $A$? Take the quadratic form and write
$$A = Q\Lambda Q^T.$$ 
(Here, I have assumed that $Q$ is orthogonal so that $Q^{-1} = Q^T$.) Substitute this decomposition of $A$ in the quadratic form and interpret what you get. Hint: you can define a new variable ${\bf y} = Q^T{\bf x}$, which is a rotated ${\bf x}$. What do we know about ${\bf y}$ given that $\Lambda$ is diagonal? What geometrical shape do we get if we plot ${\bf y}^T\Lambda{\bf y} = c$?

**Answer**: When we substitute $A = Q\Lambda Q^T$ we get our quadratic form as ${\bf x}^TQ\Lambda Q^T{\bf x}.$ If we take ${\bf y} = Q^T{\bf x}$, then ${\bf y}^T = {\bf x}^TQ$ as Q is orthogonal.

Hence our quadratic equation becomes ${\bf y}^T\Lambda {\bf y}.$ Given $\Lambda$ is diagonal and ${\bf y}$ is a rotated transformation of ${\bf x}$, we can say that the geometrical shape of the plot  ${\bf y}^T\Lambda{\bf y} = c$ will be an ellipse parralelly aligned with the new coordinate system defined by ${\bf y}$ as the equation follows the quadratic form.  

____

With that math done, let's interpret this. We wish to find outliers using a [multivariate Gaussian (MVG)](https://en.wikipedia.org/wiki/Multivariate_normal_distribution).  We model our data with the MVG and want to find all points outside of some contour - perhaps where the MVG drops to $5\%$ - all points outside of that contour are considered outliers. In a DS setting, these outlier points might trigger human intervention (e.g., to look more carefully at the credit card transaction). 

[This picture](https://en.wikipedia.org/wiki/Multivariate_normal_distribution#/media/File:MultivariateNormal.png) might be useful. 

Does the quadratic form appear naturally in the MVG?

**Answer**: The quadratic form does appear naturally in the exponential term of the MVG density function. The the matrix $A$ in our quadratic form is equal to $\Sigma^{-1}$, the inverse of covariance matrix of our variables. 

\begin{equation}\
\begin{split}
\mathcal{N}  (\mu, \Sigma) = \frac{1}{(2\pi)^{(k/2)} |\Sigma|^{1/2}} \exp \left\{- \frac{1}{2} \left(x-\mu\right)^{T} \Sigma^{-1} \left(x-\mu \right)  \right\}
\end{split}    
\end{equation}

____

Fifth question: How does the data inform the MVG? What properties of the data do we use to build the MVG? If we plot contours of the MVG, what geometric shape will they make? How is the covariance matrix of the data connected to the matrix $A$ above and what does that tell us about the contours? 

[Hint](https://geostatisticslessons.com/lessons/errorellipses). Highly recommended!! _Read very slowly and carefully...._


**Answer**: An MVG is a multivariate gaussian whose exponential term is in a quadratic form and is equal to squared of Mahalanobis distance from vector $X$ to its mean vector. An MVG of a bivariate vector whose covariance matrix is diagonal will plot a perfect bell curve in a 3D space. The MVG gives us the multivariate probability distribution of our vectors in a defined coordinate system. To build an MVG we determine an appropriate confidence interval of our data by defining the chi-squared value as per the degree of freedom and alpha. We also calculate the covariance matrix of the random vector $X$ (data). Thus, a shape of the gaussian is determined by the confidence interval and variance - covariance matrix. 

In 3D an MVG will plot a bell curve in ideal case where individual variables are not correlated. If, the variables are correlated we will see the bell getting thinner and sharper and the orientation of the shape will depend on the nature of correlation. 

In 2D an MVG will plot an hyper-ellipse whose size is determined by the diagonal elements of the covariance matrix and the shape and orientation of the ellipse is determined by the off-diagonal elements of the covariance matrix. Thus we can say that the inverse of covariance matrix in the gaussian distribution is equal to the matrix $A$ in our quadratic form.

```python
# case 1 when A is an identity matrix
A = np.array([[1,0], [0,1]])

def contour_plot(A,n_points,c,title):
    u = np.linspace(-c, c, n_points)
    x, y = np.meshgrid(u, u)

    X = np.vstack([x.flatten(), y.flatten()])
    X = X - X.mean()
    f_x = np.dot(np.dot(X.T, A), X)
    f_x = np.diag(f_x).reshape(n_points, n_points)
    plt.title(title)
    plt.contour(x, y, f_x)
    plt.show()
    

contour_plot(A,100,1,"Contour Plot")
```

```python
from mpl_toolkits.mplot3d import Axes3D 
A = np.array([[1,0], [0,1]])

def surface_plot(A,n_points,c,title):
    u = np.linspace(-c, c, n_points)
    x, y = np.meshgrid(u, u)

    X = np.vstack([x.flatten(), y.flatten()])
    X = X - X.mean()
    f_x = np.dot(np.dot(X.T, A), X) 
    f_x = np.diag(f_x).reshape(n_points, n_points)
    
    fig = plt.figure(figsize=(7,7))
    ax = plt.axes(projection='3d')
    ax.plot_surface(x,y, f_x)
    plt.title(title)
    ax.set_xlabel('X-axis', fontweight ='bold')
    ax.set_ylabel('Y-axis', fontweight ='bold')
    ax.set_zlabel('Z-axis', fontweight ='bold')
    
    plt.show()

surface_plot(A,100,1,"3d projection")
```

```python
# MVG plots
# for consistent results
np.random.seed(1)

n_samples = 125
n_outliers = 25
n_features = 2

# generate Gaussian data of shape (125, 2)
gen_cov = np.eye(n_features)
gen_cov[0, 0] = 1.
X = np.dot(np.random.randn(n_samples, n_features), gen_cov)
# add some outliers
outliers_cov = np.eye(n_features)
outliers_cov[np.arange(1, n_features), np.arange(1, n_features)] = 4.
X[-n_outliers:] = np.dot(np.random.randn(n_outliers, n_features), outliers_cov)
```

```python
my_mean = [X[:,0].mean(),X[:,1].mean()]
my_sigma = [[1,-2],[0,3]]

xlen=100
ylen=100

x_ = np.linspace(-10,10,xlen)
y_ = np.linspace(-10,10,ylen)
xx, yy = np.meshgrid(x_, y_)

def gaussian(x,y,mu,sigma):
  r = np.stack([x,y])
  inv_sigma = np.linalg.inv(sigma)
  a = 1/np.sqrt(2*np.pi)*np.sqrt(np.linalg.det(sigma))*np.exp(-((r-mu).T@inv_sigma@(r-mu))/2)

  return a


z = np.zeros((ylen,xlen))
for i in range(ylen):
  for j in range(xlen):
    z[j,i] = gaussian(x=xx[j,i], y = yy[j,i], mu= my_mean, sigma= my_sigma)
```

```python
plt.figure(figsize=(7,7))
# create a contour line at .05.
CS = plt.contour(xx,yy,z,levels=[0.05],cmap='magma_r')
plt.clabel(CS, fontsize=15,colors='black')

plt.contourf(xx,yy,z,cmap="magma_r")
plt.scatter(X[:,0],X[:,1],s=2)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```
