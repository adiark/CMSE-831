> 1. In the above code when we make a tuple of either string or integer, the conversion between data types is possible, and the data type remains constant.

````python
import numpy as np
tuple1 = ("M", list(range(5)), 1)
print(tuple1)
list1 = list(tuple1)
print(list1)
array1 = np.array(list1)
print(array1)

tuple2 = (1, "2", 3)
print(tuple2)
list2 = list(tuple2)
print(list2)
array2 = np.array(list2)
print(array2)
````
