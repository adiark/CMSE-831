Describe your kdeplot(s). How are these plots connected to summary statistics? Are the mean and standard deviation good metrics for your plots, why or why not?

📝 A Kdeplot is a density plot which tells the probability density of an attribute across its range. The area under the graph will always be equal to one. These plots on integration represents quantiles/percentiles which is what summary statistics are calculating. So in summary statistics we get the four quantile values while the KDE is a percentile plot from 0 to 100.

Mean and standard deviations are good metrics for a plot as mean tends to suggest what an average value would be in a distribution, however that is only helpful when the data point is normally distributed. If there are large number of outliers or, the data is skewed than mean is not the most accurate metric. 

Standard deviation is a very good metric as it tells the spread of the data from its mean value. This helps in understanding how wide the data is spread.

# Friday ICA: Meshgrid

In today's inclass assignment, we are going to walk through making meshgrids. Meshgrids are useful when you want to create contour plots, and 3D plots on your own. We will use them in the homework this week, so ask questions if you are having trouble!

A meshgrid creates two 2D arrays that encode a set of uniform coordinates, see the figure below.

![gog](https://i.stack.imgur.com/8Mbig.png)

In this figure, coordiantes are made for $x$ varying from $1$ to $4$ and $y$ varying from $5$ to $7$. In both $x$ and $y$, the coordinates increment by $1$. The goal is to generate the $12$ possible coordinates we can generate from $x$ and $y$. As you can see, meshgrid produces a new $XX$ array whose columns are constant and a new $YY$ array whose rows are constant. These arrays define the coordinates in the center of the figure, and are used as inputs to functions that generate contour plots and 3D images. Let's try this on our own. 

```python
# create an x array from 1 - 4, with 4 numbers
x = np.linspace(1,4,4)
# create a y array from 5 - 7, with 3 numbers
y = np.linspace(5,7,3)

print(f'x: {x}')
print(f'y: {y}')
```

```python
# create the meshgrid using numpy
XX,YY = np.meshgrid(x,y)

print(f'XX:\n {XX}')
print(f'YY:\n {YY}')
```

We have now created the coordinates for our plot, but each coordinate needs a value associated with it. For this we need to create a third array, $Z$. It will have the same shape as $XX$ and $YY$. Lets start by filling it with a constant of zeros.

```python
# create an array of zeros with the same shape as XX
z = np.zeros_like(XX)

print(f'z:\n {z}')
```

```python
# create a figure and set it to a 3D plot
fig = plt.figure(figsize=(8,5))
ax = plt.axes(projection='3d');
# pass in the coordinates and values for each coordinate
ax.plot_wireframe(XX, YY, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

Notice in our wireframe plot above that all values are set to zero, and that there are $4$ lines in the aligned with the $x$ values and $3$ lines aligned with the $y$ values of the arrays we created above. We fill in the Z array using a function.

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
```

For easy functions, you can pass the $XX$ and $YY$ arrays and automatically create $Z$.

```python
# create a figure and set it to a 3D plot
fig = plt.figure(figsize=(8,5))
ax = plt.axes(projection='3d');

z = f(XX,YY)

# pass in the coordinates and values for each coordinate
ax.plot_wireframe(XX, YY, z);
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

You also can loop through $XX$ and $YY$, to fill in $Z$.

```python
# Remember that arrays loop through the y direction first then the x direction.
# But my function takes in an x value then a y value. It can be easy to mess up.
for i in range(len(y)):
  for j in range(len(x)):
    z[i,j] = f(XX[i,j], YY[i,j])
```

```python
# create a figure and set it to a 3D plot
fig = plt.figure(figsize=(8,5))
ax = plt.axes(projection='3d');

z = f(XX,YY)

# pass in the coordinates and values for each coordinate
ax.plot_wireframe(XX, YY, z);
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

Create a surface plot (hint: `ax.plot_surface`) and filled contour plot (hint: `plt.contourf`) and use the "Reds" colormap (hint: `cmap='Reds'). Compare the two plots and how they are connected, especially the coloring. Are they telling you the same thing?

```python
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
XX,YY = np.meshgrid(x,y)
z = f(XX,YY)
```

```python
# create a figure and set it to a 3D plot
fig = plt.figure(figsize=(8,5))
ax = plt.axes(projection='3d');

z = f(XX,YY)

# make a surface plot here.
ax.plot_surface(XX,YY,z, cmap='Reds')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
# make a filled contour plot
plt.contourf(XX,YY,z,cmap='Reds')
```

Describe the plots here

The surface plot gives us a 3D image of the contour plot, as the color pattern in contour plot exactly resembles the high and low points in the surface plots. In other terms, contour plot is a top view of surface plot with color coding resembling high and low values as we go from dark red to light red.
